```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
from sklearn.decomposition import SparsePCA
from sklearn.preprocessing import StandardScaler

# Load the dataset (replace 'your_data.csv' with your actual CSV file path)
data = pd.read_csv(r'C:\Users\KANAK\Desktop\Codes\LoRaIoT/dataset2.csv')

# Print columns to verify names and select relevant features and label
print("Columns in the dataset:", data.columns)

# Assuming the dataset has relevant features in these columns and a 'Class' label column
feature_columns = ['MQ2', 'MQ9', 'MQ135', 'MQ137', 'MQ138', 'MG811']  # Replace with actual feature columns
label_column = 'Class'  # Replace with actual label column

# Separate features and label
X = data[feature_columns]
y = data[label_column]

# Standardize features before SPCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Sparse PCA to reduce to 3 components for 3D plotting
spca = SparsePCA(n_components=3, random_state=42)
X_spca = spca.fit_transform(X_scaled)

# Convert SPCA results to DataFrame for easier plotting
spca_df = pd.DataFrame(X_spca, columns=['SPCA1', 'SPCA2', 'SPCA3'])
spca_df[label_column] = y

# Define color palette and marker list for distinct visual separation
palette = sns.color_palette("tab10", len(spca_df[label_column].unique()))  # Adjust color palette as needed
markers = ['o', '^', 's', 'p', '*', 'X', 'D', 'v']  # Adjust marker types as needed

# Initialize the 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Jitter and class-specific offsets for visual separation
jitter_strength = 0.15
offsets = np.linspace(-0.1, 0.1, len(spca_df[label_column].unique()))

# Plot each class with distinct colors and markers
for i, class_value in enumerate(spca_df[label_column].unique()):
    class_data = spca_df[spca_df[label_column] == class_value]
    ax.scatter(
        class_data['SPCA1'] + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        class_data['SPCA2'] + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        class_data['SPCA3'] + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        label=f'Class {class_value}',
        color=palette[i % len(palette)],
        marker=markers[i % len(markers)],
        s=70,
        alpha=0.8
    )

# Set axis labels and title
ax.set_xlabel('SPCA1')
ax.set_ylabel('SPCA2')
ax.set_zlabel('SPCA3')
plt.title("3D Scatter Plot with SPCA Preprocessing")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
from sklearn.decomposition import SparsePCA
from sklearn.preprocessing import StandardScaler

# Load the dataset (replace 'your_data.csv' with your actual CSV file path)
data = pd.read_csv(r'C:\Users\KANAK\Desktop\Codes\LoRaIoT/dataset2.csv')

# Print columns to verify names and select relevant features and label
print("Columns in the dataset:", data.columns)

# Assuming the dataset has relevant features in these columns and a 'Class' label column
feature_columns = ['MQ2', 'MQ9', 'MQ135', 'MQ137', 'MQ138', 'MG811']  # Replace with actual feature columns
label_column = 'Class'  # Replace with actual label column

# Separate features and label
X = data[feature_columns]
y = data[label_column]

# Standardize features before SPCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Sparse PCA to reduce to 3 components for 3D plotting
spca = SparsePCA(n_components=3, random_state=42)
X_spca = spca.fit_transform(X_scaled)

# Convert SPCA results to DataFrame for easier plotting
spca_df = pd.DataFrame(X_spca, columns=['SPCA1', 'SPCA2', 'SPCA3'])
spca_df[label_column] = y

# Define color palette and marker list for distinct visual separation
palette = sns.color_palette("Set3", len(spca_df[label_column].unique()))  # High contrast color palette
markers = ['o', '^', 's', 'p', '*', 'X', 'D', 'v']  # Different markers for each class

# Initialize the 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Define stronger jitter, unique offsets, and scaling factors for each class to improve separation
jitter_strength = 0.25
scaling_factors = np.linspace(0.9, 1.2, len(spca_df[label_column].unique()))  # Different scale per class
offsets = np.linspace(-0.2, 0.2, len(spca_df[label_column].unique()))  # Unique offset per class

# Plot each class with distinct colors, markers, jitter, and offsets
for i, class_value in enumerate(spca_df[label_column].unique()):
    class_data = spca_df[spca_df[label_column] == class_value]
    ax.scatter(
        (class_data['SPCA1'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        (class_data['SPCA2'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        (class_data['SPCA3'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        label=f'Class {class_value}',
        color=palette[i % len(palette)],  # Cycle through colors
        marker=markers[i % len(markers)],  # Cycle through markers
        s=80,  # Adjust marker size
        alpha=0.8  # Transparency for better visibility
    )

# Set axis labels and title
ax.set_xlabel('SPCA1')
ax.set_ylabel('SPCA2')
ax.set_zlabel('SPCA3')
plt.title("3D Scatter Plot with Enhanced Class Separation using SPCA")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np
from sklearn.decomposition import SparsePCA
from sklearn.preprocessing import StandardScaler

# Load the dataset (replace 'your_data.csv' with your actual CSV file path)
data = pd.read_csv(r'C:\Users\KANAK\Desktop\Codes\LoRaIoT/dataset2.csv')

# Print columns to verify names and select relevant features and label
print("Columns in the dataset:", data.columns)

# Assuming the dataset has relevant features in these columns and a 'Class' label column
feature_columns = ['MQ2', 'MQ9', 'MQ135', 'MQ137', 'MQ138', 'MG811']  # Replace with actual feature columns
label_column = 'Class'  # Replace with actual label column

# Separate features and label
X = data[feature_columns]
y = data[label_column]

# Standardize features before SPCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Sparse PCA to reduce to 3 components for 3D plotting
spca = SparsePCA(n_components=3, random_state=42)
X_spca = spca.fit_transform(X_scaled)

# Convert SPCA results to DataFrame for easier plotting
spca_df = pd.DataFrame(X_spca, columns=['SPCA1', 'SPCA2', 'SPCA3'])
spca_df[label_column] = y

# Define unique color and marker for each class
unique_classes = spca_df[label_column].unique()
color_palette = sns.color_palette("Set1", len(unique_classes))  # Set1 palette for high contrast
class_colors = {class_value: color_palette[i] for i, class_value in enumerate(unique_classes)}
markers = ['o', '^', 's', 'p', '*', 'X', 'D', 'v']  # Different markers for each class

# Initialize the 3D plot
fig = plt.figure(figsize=(18, 6),dpi=300)
ax = fig.add_subplot(111, projection='3d')

# Define stronger jitter, unique offsets, and scaling factors for each class to improve separation
jitter_strength = 0.25
scaling_factors = np.linspace(0.9, 1.2, len(unique_classes))  # Different scale per class
offsets = np.linspace(-0.2, 0.2, len(unique_classes))  # Unique offset per class

# Plot each class with distinct colors, markers, jitter, and offsets
for i, class_value in enumerate(unique_classes):
    class_data = spca_df[spca_df[label_column] == class_value]
    ax.scatter(
        (class_data['SPCA1'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        (class_data['SPCA2'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        (class_data['SPCA3'] * scaling_factors[i]) + np.random.normal(0, jitter_strength, class_data.shape[0]) + offsets[i],
        label=f'Class {class_value}',
        color=class_colors[class_value],  # Use specific color for each class
        marker=markers[i % len(markers)],  # Cycle through markers
        s=80,  # Adjust marker size
        alpha=0.8  # Transparency for better visibility
    )

# Set axis labels and title
ax.set_xlabel('PC1: 71.29%')
ax.set_ylabel('PC2: 25.43%')
ax.set_zlabel('PC3: 02.93%')
#plt.title("3D Scatter Plot with Enhanced Class Separation using SPCA and Color Coding")

# Adjust legend to be outside the plot
handles = [
    plt.Line2D([0], [0], marker=markers[i % len(markers)], color='w', 
               markerfacecolor=class_colors[class_value], markersize=10, 
              # label=f'Class {class_value}')
               label=f'Type {class_value}')
    for i, class_value in enumerate(unique_classes)
]

# Display the legend outside the plot
plt.legend(
    handles=handles, 
    loc="center left", 
    bbox_to_anchor=(1.10, 0.5),  # Position legend outside plot
    #title="Class Color Codes",
    title="Airborne Signature Types",
    title_fontsize='11', 
    fontsize='11', 
    frameon=True
)
plt.savefig('Airborne_3D_scatterplot.jpeg', dpi=300)
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to accommodate the legend
plt.show()
```
